## Start a new figure
Each time call a `plt.figure()`, a new figure is created or set teh previous figure active according to num `plt.figure(num)`.

````python
x = np.arange(1, 10, 1)

plt.figure()
plt.plot(x, x**2)

plt.figure()
plt.plot(x, x**3)
````

## Set line format and axis length


```
'ro': a combination of color and string format -> read circle. Default is 'b-'

axis() takes a list of [xmin, xmax, ymain, ymax] and specify the viewpoint of the axes.

xlim(min, max)
ylim(min, max)
```

````python
plt.figure()
plt.plot([1,2,3,4], [1,4,9,16], 'ro')
plt.axis([0,6,0,20])
plt.show()

plt.figure()
plt.plot([1,2,3,4], [1,4,9,16], 'b-')
plt.xlim(0,5)
plt.ylim(0,18)
plt.show()
````

# Working with multiple figures and axes

`pyplot` has the concept of current axes and current figures. All plotting commands apply to the current axes. The function `gca()` returns the current axes and `gcf()` returns the current figure.

````python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)
````

```
plt.figure(num) -> set the cirresponding figure to active. Each time, either craete a new figure ot set the previous figure to active according to num

plt.subplot(nrows, ncols, plot_number) -> set this subplot to active, the plot_number starts from 1 to nrows*ncols
```

````python
plt.figure(1)                # the first figure

plt.subplot(211)             # the first subplot in the first figure, set 211 to active
plt.plot([1, 2, 3])

plt.subplot(212)             # the second subplot in the first figure, set 222 to active
plt.plot([4, 5, 6])

plt.figure(2)                # a second figure
plt.plot([4, 5, 6])          # creates a subplot(111) by default

plt.figure(1)                # figure 1 current; subplot(212) still current
plt.subplot(211)             # make subplot(211) in figure1 current
plt.title('Easy as 1, 2, 3') # subplot 211 title
````

## Memory release

If you are making lots of figures, you need to be aware of one more thing: **the memory required for a figure is not completely released until the figure is explicitly closed with `close()`**. Deleting all references to the figure, and/or using the window manager to kill the window in which the figure appears on the screen, is not enough, because pyplot maintains internal references until close() is called.

To close all the figure
```
plt.close('all')
```

However, when using **jupyter notebook**, I think there is a call for `close()` after running the code block. 

````python
plt.figure(1)                # figure 1 current; subplot(212) still current
plt.subplot(212)             # make subplot(211) in figure1 current
plt.title('Easy as 1, 2, 3') # subplot 211 title

# there is no plot on the figure, memory is released.
````

# Working with text

The `text()` command can be used to add text in an arbitrary location, and the `xlabel()`, `ylabel()` and `title()` are used to add text in indicated locations.

matplotlib accepts TeX equation expressions in any text expression:
```
r'$\mu=100,\ \ sigma=15$'
```

````python
# Fixing random state for reproducibility
np.random.seed(19680801)

mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

n, bins, patchs = plt.hist(x, 50, normed=1, facecolor='g', alpha=0.75)

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \ sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
````

## Annotating text

````python
t = np.arange(.0, 5.0, 0.01)
s = np.cos(2 * np.pi * t)

plt.plot(t, s, lw=2)

plt.annotate('local max', xy=(2,1), xytext=(3, 1.5),
            arrowprops=dict(facecolor='black', shrink=0.05))

plt.ylim(-2,2)
plt.show()
````

# Logarithmic and other nonlinear axes

Changing the scale of an axis is easy:
```
plt.xscale('log')
plt.yscale('log')
```

````python
from matplotlib.ticker import NullFormatter
````

````python
# Fixing random state for reproducibility
np.random.seed(19680801)

y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y>0) & (y<1)]
y.sort()
x = np.arange(len(y))
````

````python
# linear
plt.figure()
plt.subplot(211)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)

# log
plt.subplot(212)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)
````
