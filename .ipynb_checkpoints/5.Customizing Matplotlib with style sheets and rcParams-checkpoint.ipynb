````python
plt.style.use('ggplot')
data = np.random.randn(50)
plt.plot(data)
````

## Temporary styling

If you only want to **use a style for a specific block of code** but don't want to change the global styling, the style package provides a context manager for limiting your changes to a specific scope.

````python
plt.figure(figsize=(10, 6))
with plt.style.context('dark_background'):
    plt.plot(np.sin(np.linspace(0, 2*np.pi)), 'r-o')
plt.show()
````

# matplotlib rcParams

## Dynamic rc settings

You can dynamically change the default rc settings in a python script or interactively from the python shell. All of the rc settings are stored in a dictionary-like variable called `matplotlib.rcParams`.

````python
mpl.rcParams['lines.linewidth'] = 5
plt.plot(data)
````

The `matplotlib.rc()` command can be used to modify multiple settings in a single group at once, using keyword arguments

The `matplotlib.rcdefaults()` command will restore the standard matplotlib default settings.

````python
mpl.rc('lines', linewidth=4, color='g')
plt.plot(data)
````

````python
mpl.rcdefaults()
plt.plot(data)
````

## The `matplotlibrc` file

matplotlib uses matplotlibrc configuration files to **customize all kinds of properties**, which we call rc settings or rc parameters. You can control the defaults of almost every property in matplotlib: figure size and dpi, line width, color and style, axes, axis and grid properties, text and font properties and so on.

To display where the currently active matplotlibrc file was loaded from, one can do the following:

````python
mpl.matplotlib_fname()
````

By checking the `matplotlibrc` file, your can get which params that you can customize in your code
